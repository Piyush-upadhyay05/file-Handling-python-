using DataFrame.to_csv() ways to write a csv file using pandas 

```python
import pandas as pd

data = {
    'Name': ['Piyush', 'Anita'],
    'Age': [21, 20],
    'Grade': ['A', 'B']
}

df = pd.DataFrame(data)
df.to_csv('staff.csv', index=False)  # index=False avoids writing row numbers
```

quoting and escaping

Quoting 

1. QUOTE_ALL

```python
import csv

data = [['Name', 'Quote'],
        ['Piyush', 'He said, "I will succeed"'],
        ['Anita', 'Smart, Honest']]

with open('quotes.csv', mode='w', newline='') as file:
    writer = csv.writer(file, quoting=csv.QUOTE_ALL)
    writer.writerows(data)
```

| Constant               | Meaning                                                                 |
| ---------------------- | ----------------------------------------------------------------------- |
| `csv.QUOTE_MINIMAL`    | (default) Only quote fields if needed (e.g., contains comma or newline) |
| `csv.QUOTE_ALL`        | Quote **every field**                                                   |
| `csv.QUOTE_NONNUMERIC` | Quote all non-numeric fields                                            |
| `csv.QUOTE_NONE`       | Don’t quote anything (you must use escaping instead)                    |


Escaping

```python
import csv

data = [['Name', 'Comment'],
        ['Piyush', 'Topper, "intelligent"'],
        ['Anita', 'A+, "coder"']]

with open('escaped.csv', mode='w', newline='') as file:
    writer = csv.writer(file, quoting=csv.QUOTE_NONE, escapechar='\\')
    writer.writerows(data)
```

reading csv both quoting and escaping 

```python
with open('quotes.csv', mode='r') as file:
    reader = csv.reader(file, quoting=csv.QUOTE_ALL)
    for row in reader:
        print(row)
```

or for escape 

```python
with open('escaped.csv', mode='r') as file:
    reader = csv.reader(file, quoting=csv.QUOTE_NONE, escapechar='\\')
    for row in reader:
        print(row)
```
