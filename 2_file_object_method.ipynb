file object method

```python
f = open("data1.txt",mode='r')
print(f.readable())
print(f.writable())
f.close()
```

```python
f = open("data1.txt",mode='w')
print(f.readable())
print(f.writable())
f.close()
```

```python
f = open("data1.txt",mode='r+') # r+/w+ means purpose of this file mode is to read and write both 
print(f.readable())
print(f.writable())
f.close()
```

| Method / Property | Description                    | Returns          |
| ----------------- | ------------------------------ | ---------------- |
| `readable()`      | Can we read from file?         | `True` / `False` |
| `writable()`      | Can we write to file?          | `True` / `False` |
| `closed`          | Is file closed?                | `True` / `False` |
| `isatty()`        | Is file connected to terminal? | `True` / `False` |

to remove this error 

```python
import os
if os.path.isfile("hello.txt"):
    f = open("hello.txt")
    # operations 
    f.close()
else:
    print("file does not exist")

```

taking file name input from the user 

```python
import os

filename = input("enter your file name")
if os.path.isfile(filename):
    f = open(filename)
    # oeprations 
    f.close()
    print("file exist")
else:
    print("file does not exist")
```

| Method | Function Used      | Remarks                               |
| ------ | ------------------ | ------------------------------------- |
| 1️⃣    | `os.path.exists()` | Checks if path exists (file/folder)   |
| 2️⃣    | `os.path.isfile()` | Confirms it's a file, not a directory |
| 3️⃣    | `Path().exists()`  | Cleaner, modern approach              |
| 4️⃣    | `try...except`     | Checks and handles missing file       |
| 5️⃣    | `os.access()`      | Checks existence + permissions        |


mode of opening text file 


| Mode    | Type   | Description              |
| ------- | ------ | ------------------------ |
| `'r'`   | Text   | Read only                |
| `'w'`   | Text   | Write (overwrite)        |
| `'a'`   | Text   | Append at end            |
| `'x'`   | Text   | Create new file          |
| `'r+'`  | Text   | Read + Write             |
| `'w+'`  | Text   | Write + Read (overwrite) |
| `'a+'`  | Text   | Append + Read            |
| `'x+'`  | Text   | Create + Read/Write      |



following are mode for binary file mode 

| Mode    | Type   | Description              |
| ------- | ------ | ------------------------ |
| `'rb'`  | Binary | Read in binary           |
| `'wb'`  | Binary | Write in binary          |
| `'ab'`  | Binary | Append in binary         |
| `'xb'`  | Binary | Create new file (binary) |
| `'rb+'` | Binary | Read + Write in binary   |
| `'wb+'` | Binary | Write + Read in binary   |
