```python
files = []
merged_data = ""

while True:
    f_name = input("Enter the file name (without .txt): ")
    files.append(f_name)
    ans = input("Do you want to add another file? (y/n): ").lower()
    if ans != 'y':
        break

for file in files:
    filename = file + '.txt'
    f = open(filename, mode='r')
    merged_data = merged_data + f.read() + '\n'
    f.close()

print(merged_data)
with open(input("enter the final file name")+'.txt', mode='x') as f:
    f.write(merged_data)
    print("merging is done thanks")
```

```python
# suppose you write a file name which is not present in directory for this 
import os 
files = []
merged_data = ""

while True:
    f_name = input("Enter the file name (without .txt): ")
    files.append(f_name)
    ans = input("Do you want to add another file? (y/n): ").lower()
    if ans != 'y':
        break

for file in files:
    filename = file + '.txt'
    if os.path.isfile(filename):
       f = open(filename, mode='r')
       merged_data = merged_data + f.read() + '\n'
       f.close()

print(merged_data)
with open(input("enter the final file name")+'.txt', mode='x') as f:
    f.write(merged_data)
    print("merging is done thanks")
```
