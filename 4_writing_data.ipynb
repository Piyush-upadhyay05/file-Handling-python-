writing data to a text file 

```python
# using write method 
f = open('write_method.txt', mode='w')
f.write("hello\nworld") 
n = f.write("bye bye") # be aware the cursor position 
print(n) # write also use to print the number of char is written 
f.close()
```

```python
# using writelines method 
f = open('writelines.txt', mode='w')
line_lists= ['yash\n','yashtii\n','chiku\n','ohayo\n']
f.writelines(line_lists)
f.close()
```

What is writelines()?

 -   The writelines() method is used to write a sequence (like a list or iterable) of strings to a file.

 -    It does not return anything, meaning the return value is None.

  -  It simply writes and finishes silently.

copy content of one file to another 

```python
# making file name first.txt with some data in it 
f = open("first.txt", mode='w')
f.write("hello\nmy name is piyush upadhyay\none day i become my dreams true")
f.close()
```

```python
# making a empty file 
f = open("second.txt", mode='w')
f.close()
```

```python
# copy content of first.txt to second.txt 
f1 = open("first.txt", mode='r')
f2 = open("second.txt", mode='w')
data = f1.readlines()
for line in data:
    f2.write(line)
f1.close()
f2.close()
# now you see the content of first.txt data is copy to the second.txt
```

```python
# you can also use this to copy content  
f1 = open("first.txt", mode='r')
f2 = open("second.txt", mode='w')
for line in f1:
    f2.write(line)
f1.close()
f2.close()
```
