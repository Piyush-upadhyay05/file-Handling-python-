wrting json to a files using json.dump()

```python
import json

# Python dictionary (data to be saved)
student_data = {
    "name": "Piyush",
    "age": 21,
    "skills": ["Python", "AI", "ML"],
    "is_student": True
}

# Write the dictionary to a JSON file
with open("student.json", "w") as file:
    json.dump(student_data, file, indent=4)
```

convert python to a json string using json.dumps()

```python
import json

data = {
    "subject": "Machine Learning",
    "level": "Intermediate"
}

json_str = json.dumps(data, indent=2)
print(json_str)
```

parameters in json.dump() and json.dumps()

| Parameter                | Description                             |
| ------------------------ | --------------------------------------- |
| `indent=4`               | Pretty prints the JSON with indentation |
| `sort_keys=True`         | Sorts dictionary keys alphabetically    |
| `separators=(',', ': ')` | Customize how items are separated       |

- Use indent and sort_keys to make JSON readable.


Example with all options:

```python
json.dump(data, file, indent=4, sort_keys=True, separators=(",", ": "))
```
