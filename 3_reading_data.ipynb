reading data from a file

- 1. read()

```python
f = open("intro.txt",'r')
data = f.read()
print(data)
f.close()
```

```python
f = open("intro.txt",'r')
data = f.read(2)
# f = open("intro.txt",'r')
data1 = f.read(7)
print(data)
print(data1)
f.close()
```

```python
f = open("intro.txt",'r')
data = f.read(-4)
print(data)
f.close()
```

- 2. readline()

```python
# Create and write 3 lines into a text file
with open("sample.txt", "w") as f:
    f.write("This is line 1\n")
    f.write("This is line 2\n")
    f.write("This is line 3\n")
```

```python
f = open("sample.txt",'r')
data = f.readline()
data1 = f.readline()
print(data)
print(data1)
f.close()
```

remove space between two line in output just use end 


```python
f = open("sample.txt",'r')
data = f.readline()
data1 = f.readline()
print(data,end="")
print(data1,end="")
f.close()
```

```python
f = open("sample.txt",'r')
data = f.readline(3)
data1 = f.readline(2)
print(data)
print(data1)
f.close()
```

readlines()

```python
f = open("sample.txt",'r')
data = f.readlines()
print(data)
for i in data:
    print(i,end="")
f.close()
```

reading file using for loop 

```python
f = open("sample.txt",'r')
for line in f:
    print(line,end="")
f.close()
```

file handling two important methods 

- 1. tell() method 
- 2. seek() method

tell method

```python
f = open("sample.txt",mode='r')
print(f.tell())
f.read(3)
print(f.tell()) 
f.read()
print(f.tell())
```

seek() method

```python
f = open("sample.txt", 'r')
print(f.tell())
print(f.read(3))
f.seek(0)
print(f.tell())
print(f.read())
f.close()
```

finding number of words and char and line 

```python
f= open('sample.txt',mode='r')
number_of_char = 0
number_of_words = 0
number_of_lines = 0

for line in f:
    number_of_lines+=1
    line = line.strip("\n")
    number_of_char+=len(line)
    list1 = line.split()
    number_of_words+=len(list1)
f.close()
print("number of line is :",number_of_lines)
print("number of char is :",number_of_char)
print("number of words is :", number_of_words)
```
