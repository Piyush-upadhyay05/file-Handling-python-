readiung from josn string 

```python
import json

# JSON string (text format)
json_data = '{"name": "Piyush", "age": 21, "is_student": true}'

# Convert JSON string to Python dictionary
data = json.loads(json_data)

print(data)
print(data['name'])   # Access individual value
```

reading from json file 

```python
import json

# Open and read JSON file
with open("example.json", "r") as file:
    data = json.load(file)

# Now 'data' is a Python dictionary
print(data)
print(data["skills"])  # Access list inside JSON
```
