renaming multiple files in python 

```python
import os 

# rename(old_name,new_name)
path = input("entre the path")
path = path.replace('\\','/')
print(path)
print(os.listdir(path))

def main():
    i = 1
    for filename in os.listdir(path):
        new_name = path+"car"+str(i)+'.jpg'
        old_name = path+filename
        os.rename(old_name,new_name)
        i+=1
    
main()
```

file object attributes 

| Attribute     | Description                                                                    |
| ------------- | ------------------------------------------------------------------------------ |
| `.name`       | Returns the **name of the file** (as passed to `open()`).                      |
| `.mode`       | Returns the **mode** in which the file was opened (`'r'`, `'w'`, `'rb'`, etc.) |
| `.closed`     | Returns `True` if the file is **closed**, else `False`.                        |
| `.readable()` | Returns `True` if the file can be **read from**.                               |
| `.writable()` | Returns `True` if the file can be **written to**.                              |
| `.seekable()` | Returns `True` if the file supports **random access** (seeking).               |
| `.encoding`   | Returns the encoding used to decode or encode the file (like `'UTF-8'`).       |
| `.fileno()`   | Returns the file’s **file descriptor** (an integer).                           |
| `.flush()`    | Flushes the internal buffer (write contents to disk immediately).              |
| `.isatty()`   | Returns `True` if the file is connected to a **TTY device** (console).         |
| `.tell()`     | Returns the current **file cursor position**.                                  |


```python
file = open("data.txt", "r")

print("File name:", file.name)
print("File mode:", file.mode)
print("Is file closed?:", file.closed)
print("Is file readable?:", file.readable())
print("Is file writable?:", file.writable())
print("Is file seekable?:", file.seekable())

file.close()
print("After closing, is file closed?:", file.closed)
```
