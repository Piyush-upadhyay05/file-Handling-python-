##### 1. To read csv files 

- 1. in form of list 

```python
import csv

with open("titanic.csv",'r') as f:
    titanic_data = csv.reader(f,delimiter=',')
    print(titanic_data)

    for data in titanic_data:
        print(data)
```

- 2. in form of dictionary 

```python
import csv

with open("titanic.csv",'r') as f:
    titanic_data = csv.DictReader(f,delimiter=',')
    print(titanic_data)

    for data in titanic_data:
        print(data)



```
