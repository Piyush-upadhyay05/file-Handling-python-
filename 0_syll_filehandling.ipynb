### ✅ **1. Introduction to File Handling**

* **What is File Handling?**

  * Managing files (read, write, append, etc.) using Python.
  * Files store data permanently.

* **Types of Files**

  * **Text files** (`.txt`, `.csv`, `.log`, etc.)
  * **Binary files** (`.jpg`, `.exe`, `.mp4`, `.dat`, etc.)

---

### ✅ **2. File Operations in Python**

#### 📌 **Basic Steps of File Handling**

1. Open a file
2. Perform operations (read/write/append)
3. Close the file

---

### ✅ **3. Opening Files**

```python
file = open("filename.txt", "mode")
```

#### 📌 **File Modes**

| Mode  | Description                                       |
| ----- | ------------------------------------------------- |
| `'r'` | Read (default) – file must exist                  |
| `'w'` | Write – creates a new file or overwrites existing |
| `'a'` | Append – adds data to end of file                 |
| `'x'` | Exclusive creation – fails if file exists         |
| `'b'` | Binary mode – e.g., `'rb'`, `'wb'`                |
| `'t'` | Text mode – default                               |

#### 🔍 Examples:

```python
open("data.txt", "r")     # read mode
open("data.txt", "w")     # write mode
open("data.txt", "a")     # append mode
open("image.jpg", "rb")   # read binary
```

---

### ✅ **4. Reading from a File**

#### 📌 **Methods to Read**

| Method        | Description                   |
| ------------- | ----------------------------- |
| `read()`      | Reads entire file as a string |
| `read(n)`     | Reads first `n` characters    |
| `readline()`  | Reads one line                |
| `readlines()` | Reads all lines into a list   |

```python
f = open("demo.txt", "r")
print(f.read())           # Entire file
print(f.readline())       # First line
print(f.readlines())      # List of lines
f.close()
```

---

### ✅ **5. Writing to a File**

#### 📌 **Methods**

| Method         | Description              |
| -------------- | ------------------------ |
| `write()`      | Writes a string          |
| `writelines()` | Writes a list of strings |

```python
f = open("demo.txt", "w")
f.write("Hello, World!\n")
f.writelines(["Line 1\n", "Line 2\n"])
f.close()
```

---

### ✅ **6. Appending to a File**

```python
f = open("demo.txt", "a")
f.write("Appended content.\n")
f.close()
```

---

### ✅ **7. Closing a File**

* **Always close the file to free system resources**

```python
f.close()
```

---

### ✅ **8. Using `with` Statement (Context Manager)**

* **Automatically closes the file**

```python
with open("demo.txt", "r") as f:
    content = f.read()
```

---

### ✅ **9. Working with Binary Files**

#### 🔍 Read/Write binary

```python
with open("pic.jpg", "rb") as f:
    data = f.read()

with open("copy.jpg", "wb") as f:
    f.write(data)
```

---

### ✅ **10. File Object Attributes**

| Attribute     | Description                    |
| ------------- | ------------------------------ |
| `file.name`   | File name                      |
| `file.mode`   | Mode used                      |
| `file.closed` | Boolean: file is closed or not |

```python
with open("test.txt", "r") as f:
    print(f.name)
    print(f.mode)
    print(f.closed)
print(f.closed)  # True
```

---

### ✅ **11. File Pointer Operations**

| Function               | Description                    |
| ---------------------- | ------------------------------ |
| `tell()`               | Current position of pointer    |
| `seek(offset, whence)` | Move pointer to a new position |

#### 🔍 Example:

```python
f = open("demo.txt", "r")
print(f.tell())          # 0
print(f.read(5))         # Reads 5 chars
print(f.tell())          # 5
f.seek(0)                # Back to start
f.close()
```

### ✅ **12. Handling CSV Files**

Use the `csv` module.

```python
import csv

# Writing CSV
with open('data.csv', 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(["name", "age"])
    writer.writerow(["Alice", 25])

# Reading CSV
with open('data.csv', 'r') as file:
    reader = csv.reader(file)
    for row in reader:
        print(row)
```

---

### ✅ **13. Handling JSON Files**

Use the `json` module.

```python
import json

# Write JSON
data = {"name": "Alice", "age": 25}
with open("data.json", "w") as f:
    json.dump(data, f)

# Read JSON
with open("data.json", "r") as f:
    data = json.load(f)
    print(data)
```

### ✅ **14. File Exception Handling**

```python
try:
    with open("demo.txt", "r") as f:
        content = f.read()
except FileNotFoundError:
    print("File not found")
except IOError:
    print("Input/output error")
```

### ✅ **15. Advanced Topics**

#### 🔸 Working with `os` module

```python
import os
print(os.getcwd())         # Current directory
print(os.listdir())        # List files
os.rename("a.txt", "b.txt") # Rename file
os.remove("b.txt")          # Delete file
```

#### 🔸 Working with `pathlib` (modern way)

```python
from pathlib import Path
file = Path("demo.txt")
if file.exists():
    print(file.read_text())
```
