reading csv file using pandas.read_csv()

```python
import pandas as pd

df = pd.read_csv('titanic.csv')
print(df)
```

In using pd.read_csv() commonly used parameters are following 
| Parameter            | Purpose                            |
| -------------------- | ---------------------------------- |
| `sep` or `delimiter` | Change separator (default: `,`)    |
| `header`             | Row to use as column names         |
| `names`              | Provide custom column names        |
| `index_col`          | Set a column as index              |
| `usecols`            | Read specific columns only         |
| `nrows`              | Read only first N rows             |
| `skiprows`           | Skip initial rows                  |
| `na_values`          | Treat certain strings as NaN       |
| `dtype`              | Set data types manually            |
| `parse_dates`        | Convert date columns automatically |

1. delimiter/sep

```python
import pandas as pd

df = pd.read_csv('titanic.csv')
df = pd.read_csv('titanic.csv',sep=';')
print(df)
```

2. header

```python
import pandas as pd

df = pd.read_csv('titanic.csv')
df = pd.read_csv('titanic.csv', header=None)
print(df)
```

3. names

```python
import pandas as pd 
df = pd.read_csv('titanic.csv', names=['ID', 'Name', 'Gender', 'Seat', 'Surname'], header=0)
print(df)
```

4. index_col

```python
import pandas as pd
df = pd.read_csv('titanic.csv', index_col='Passenger id')

print(df)
```

5. usecols

```python
import pandas as pd 
df = pd.read_csv('titanic.csv', usecols=['name', 'seat number'])
print(df)
```

6. nrows

```python
import pandas as pd
df = pd.read_csv('titanic.csv', nrows=2)
print(df)
```

7. skiprows

```python
import pandas as pd 
df = pd.read_csv('titanic.csv', skiprows=1)
print(df)
```

8. na_values

```python
import pandas as pd
df = pd.read_csv('titanic.csv', na_values=['-1'])
print(df)
```

9. dtype 

```python
import pandas as pd 
df = pd.read_csv('titanic.csv', dtype={'Passenger id': float, 'seat number': int})
print(df)
```

10. parse_dates

```python
import pandas as pd
df = pd.read_csv('titanic.csv', parse_dates=['timestamp'])
print(df)
```

fetch data as a function 

```python
import csv 
def check_csv(file):
    with open(file,mode='r') as f:
        my_data = csv.reader(f,delimiter=',')
        header = next(my_data)
        print(header)
        print(len(header))
check_csv('titanic.csv')
```

python program to check csv format is valid or invalid 

```python
import csv 
def check_csv(file):
    with open(file,mode='r') as f:
        my_data = csv.reader(f,delimiter=',')
        header = next(my_data)
        number_of_columns = len(header)
        for rec in my_data:
            if len(rec)!=number_of_columns:
                return False
        return True
detector = check_csv('titanic.csv')
if detector == True:
    print("valid format")
else:
    print("invalid format")
```
