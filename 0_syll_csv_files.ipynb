### ✅ 4. **File Opening Modes for CSV**

* `"r"` – Read mode
* `"w"` – Write (overwrites file)
* `"a"` – Append (adds at end)
* `"r+"` – Read and Write
* `"newline=''`" to avoid blank lines in Windows
