there are two ways to write a csv file

- 1. csv module 
- 2. using pandas 

1 . using csv module

- i. using csv.writer()
  - This method writes rows (lists) into a CSV file. 

```python
import csv

with open('students.csv', mode='w', newline='') as f:
    writer = csv.writer(f)
    writer.writerow(['Name', 'Age', 'Grade'])  # Header
    writer.writerow(['Piyush', 21, 'A'])       # Row 1
    writer.writerow(['Anita', 20, 'B'])        # Row 2
```

- ii. using csv.DictWriter()
  - This method writes rows using dictionaries with keys as column names. 


```python
import csv

with open('teacher.csv', mode='w', newline='') as file:
    fieldnames = ['Name', 'Age', 'Grade']
    writer = csv.DictWriter(file, fieldnames=fieldnames)

    writer.writeheader()  # Write header row
    writer.writerow({'Name': 'Piyush', 'Age': 21, 'Grade': 'A'})
    writer.writerow({'Name': 'Anita', 'Age': 20, 'Grade': 'B'})
```
