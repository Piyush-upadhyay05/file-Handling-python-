converting json to csv

```python
# example json data which we have to convert 
data = [
    {"name": "Piyush", "age": 21, "city": "Delhi"},
    {"name": "Aarav", "age": 22, "city": "Mumbai"},
    {"name": "Riya", "age": 20, "city": "Pune"}
]
```

```python
import json
import csv

# Data to convert
data = [
    {"name": "Piyush", "age": 21, "city": "Delhi"},
    {"name": "Aarav", "age": 22, "city": "Mumbai"},
    {"name": "Riya", "age": 20, "city": "Pune"}
]

# Write to CSV file
with open("conversion.csv", "w", newline="") as file:
    writer = csv.DictWriter(file, fieldnames=["name", "age", "city"])
    writer.writeheader()  # Write column names
    writer.writerows(data)
```

converting csv to json 

```python
# example data to convert to csv 
name,age,city
Piyush,21,Delhi
Aarav,22,Mumbai
Riya,20,Pune
```

```python
import csv
import json

# Read from CSV
with open("conversion.csv", "r") as file:
    reader = csv.DictReader(file)
    data = list(reader)

# Convert to JSON and save
with open("conversing.json", "w") as json_file:
    json.dump(data, json_file, indent=4)
```
