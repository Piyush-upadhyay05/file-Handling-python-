#### python json module 

- 1. json.load/json.loads() = convert json to python 
- 2. json.dump/json.dumps() = convert pyhton to json 

1. json.dumps()

```python
import json

data = {"name": "Piyush", "age": 21, "is_student": True}
json_string = json.dumps(data)
print(json_string)
```

2. json.dump()

```python
data = {"name": "Piyush", "skills": ["Python", "ML"]}

with open("data.json", "w") as f:
    json.dump(data, f, indent=2)
```

3. json.loads()

```python
json_str = '{"name": "Piyush", "age": 21}'
data = json.loads(json_str)
print(data["name"])  # Output: Piyush
```

4. json.load()

```python
with open("data.json", "r") as f:
    data = json.load(f)

print(data)
```
