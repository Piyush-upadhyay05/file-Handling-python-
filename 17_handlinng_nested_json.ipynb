example of nested json

```python
{
  "student": {
    "name": "Piyush",
    "age": 21,
    "address": {
      "city": "Delhi",
      "pincode": 110001
    },
    "skills": ["Python", "ML"]
  }
}
```

converting this into python dictionary 

```python
data = {
  "student": {
    "name": "Piyush",
    "age": 21,
    "address": {
      "city": "Delhi",
      "pincode": 110001
    },
    "skills": ["Python", "ML"]
  }
}
```

accessing the value from nested json

```python
import json

# JSON string (you can also read from file using json.load())
json_data = '''
{
  "student": {
    "name": "Piyush",
    "age": 21,
    "address": {
      "city": "Delhi",
      "pincode": 110001
    },
    "skills": ["Python", "ML"]
  }
}
'''

# Convert JSON string to Python dict
data = json.loads(json_data)

# Access nested values
print(data["student"]["name"])         # Output: Piyush
print(data["student"]["address"]["city"])  # Output: Delhi
print(data["student"]["skills"][0])    # Output: Python
```

reading nested json from file 

```python
with open("student.json", "r") as file:
    data = json.load(file)

print(data["student"]["address"]["pincode"])  # Output: 110001
```
