## what is file ?

- File are named locations on a disk to store informations 

- they are used to store data permanently 

- Data is stored in non-volatile memory 

   - Non-volatile memory means:

      - It does not lose data when the computer is turned off.

      - Examples: Hard Disk, SSD, USB Drive, SD Card 

- we can retrieve data whenever we required  

```python
age = str(input("enter your age"))
f = open("data.txt","w")
f.write(age)
f.close()
```

to read the file 

```python
f = open("data.txt",mode='r')
print(f.read())
f.close()
```

```python
f = open("data.txt", 'r')
# operations
print(f.read())
f.close()
# after closing you cant perform any operations in it 
```

file object variable 

```python
f = open("data1.txt",'r')
print("file name is ",f.name)
print("encoding is ",f.encoding)
print("file purpose is ",f.mode)
print("is file is closed",f.closed)
```
