# 📘 What Is a JSON File?

**JSON (JavaScript Object Notation)** is a lightweight **data-interchange format**, easy for **humans to read** and **machines to parse**.

🔹 It stores **data as key-value pairs**, similar to Python dictionaries.
🔹 Commonly used for data exchange between **server and client**.

---

## ✅ JSON File Example:

```json
{
  "name": "Piyush",
  "age": 21,
  "skills": ["Python", "ML", "Data Analysis"],
  "is_student": true
}
```

## 🔸 2. **Python `json` Module Basics**

### 🔹 Common Functions:

| Function       | Description                                   |
| -------------- | --------------------------------------------- |
| `json.load()`  | Reads JSON from a file and converts to Python |
| `json.loads()` | Converts JSON string to Python                |
| `json.dump()`  | Writes Python object to a JSON file           |
| `json.dumps()` | Converts Python object to JSON string         |

---

## 🔸 3. **Reading JSON in Python**

### ✅ Example using `json.load()`:

```python
import json

with open('data.json', 'r') as file:
    data = json.load(file)
    print(data)
```

### ✅ Example using `json.loads()`:

```python
json_string = '{"name": "Piyush", "age": 21}'
data = json.loads(json_string)
print(data['name'])  # Output: Piyush
```

---

## 🔸 4. **Writing JSON in Python**

### ✅ Example using `json.dump()`:

```python
import json

data = {'name': 'Piyush', 'age': 21, 'skills': ['Python', 'ML']}

with open('data.json', 'w') as file:
    json.dump(data, file)
```

### ✅ Example using `json.dumps()`:

```python
json_string = json.dumps(data)
print(json_string)
```

---

## 🔸 5. **JSON Data Types vs Python Data Types**

| JSON    | Python       |
| ------- | ------------ |
| Object  | dict         |
| Array   | list, tuple  |
| String  | str          |
| Number  | int, float   |
| Boolean | True / False |
| Null    | None         |

---

## 🔸 6. **Pretty Printing JSON**

Use `indent` and `sort_keys` to make JSON readable.

```python
print(json.dumps(data, indent=4, sort_keys=True))
```

---

## 🔸 7. **Handling Nested JSON Objects**

```json
{
  "name": "Piyush",
  "education": {
    "college": "XYZ University",
    "year": 2025
  }
}
```

Access in Python:

```python
print(data['education']['college'])  # Output: XYZ University
```

---

## 🔸 8. **Updating JSON Data**

```python
data['age'] = 22
data['skills'].append('Deep Learning')
```

---

## 🔸 9. **Converting Between JSON and CSV (Bonus)**

Sometimes you need to convert JSON ↔ CSV.

Using pandas:

```python
import pandas as pd
df = pd.read_json('data.json')
df.to_csv('data.csv', index=False)
```

## 🔸 10. **Common Errors in JSON Handling**

* Invalid syntax (e.g., single quotes instead of double quotes)
* Trailing commas
* Incorrect data types (e.g., using Python's `None` in JSON)
* Encoding issues (`utf-8`)
