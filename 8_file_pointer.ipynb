file pointer operations on python 

- 1. tell()
- 2. seek(offset, whence)

tell()

```python
f = open("second.txt", "r")
print("Initial position:", f.tell())  # Usually 0
data = f.read(5)
print("After reading 5 characters:", f.tell())
f.close()
```

seek(offset,whence)

```python
f = open("second.txt", "r")

f.seek(10)         # Move to 10th byte from beginning
print(f.tell())    # Should print 10

f.seek(0, 2)       # Move to end of file
print(f.tell())    # Position at end (file size in bytes)

f.close()
```
